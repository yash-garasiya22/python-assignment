```python
# Q:5] create a list and tuple in Python, and demonstrate how attempting to change an element works differently

# list (mutable)
my_list = [10, 20, 30, 40, 50]
print("Original list:", my_list)
my_list[0] = 100
print("Modified list:", my_list)

# Creating a tuple (uses parentheses)
my_tuple = (10, 20, 30, 40, 50)
print("\nOriginal tuple:", my_tuple)
# Attempting to modify a tuple element will raise a TypeError
try:
    my_tuple[0] = 100
except TypeError as e:
    print(f"Attempting to modify tuple element raised an error: {e}")
```

```python
# Q:13] Write a Python program to find the sum of all even numbers between 1 and 50?

sum_even = 0

for i in range (1,50):
    if i % 2 == 0:
        sum_even += i
        print(sum_even)



print(f"The sum of all even numbers between 1 and 50 is: {sum_even}")

```

```python
# Q:14]  Write a program to reverse a string using a while loop.

row = "tendulkar"
row1 = ""
i = len(row) - 1

while i >= 0:
    row1 += row[i]
    i -= 1
print(row1)
```

```python
# Q:15] Write a Python program to calculate the factorial of a number provided by the user using a while loop?

num = int(input("Enter a number: "))
factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print(f"The factorial of {num} is {factorial}")
```
